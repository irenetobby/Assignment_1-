Bank Account Manager Using Python Basics

```python
# starting balance
balance = 1000

def deposit(balance, amount):
    balance = balance + amount
    print(f"Deposited ₹{amount}. New Balance: ₹{balance}")
    return balance

def withdraw(balance, amount):
    if amount <= balance:
        balance = balance - amount
        print(f"Withdrew ₹{amount}. Remaining Balance: ₹{balance}")
    else:
        print("Insufficient balance! Withdrawal failed.")
    return balance

def check_balance(balance):
    print(f"Current Balance: ₹{balance}")
```

```python
balance = deposit(balance, 500)
balance = withdraw(balance, 200)
check_balance(balance)

balance = withdraw(balance, 2000)   # this will fail because no balance
check_balance(balance)
```

```python
balance = withdraw(balance, 100)   # this will fail because no balance
check_balance(balance)
balance=withdraw(balance,300)
check_balance(balance)
```

Weather Data Analyzer

```python

# Weather data for a region
region = "Mumbai"
temperatures = [30, 32, 31, 29, 35, 36, 33]        # temperature in °C
humidity_levels = [75, 80, 78, 70, 85, 90, 82]     # humidity in %

# Function to determine weather condition
def classify_weather(temp, humidity):
    if temp > 33:
        return "Hot"
    elif humidity > 85:
        return "Possible Rain"
    else:
        return "Normal"

# Function to calculate average temperature
def average_temperature(temp_list):
    return sum(temp_list) / len(temp_list)

# Function to calculate average humidity
def average_humidity(humidity_list):
    return sum(humidity_list) / len(humidity_list)

# Function to analyze day-wise weather
def analyze_daily_weather(temps, humidity):
    print(f"Weather Report for {region}:\n")
    for i in range(len(temps)):
        condition = classify_weather(temps[i], humidity[i])
        print(f"Day {i+1}: Temp = {temps[i]}°C, Humidity = {humidity[i]}%, Condition = {condition}")

# Display report
analyze_daily_weather(temperatures, humidity_levels)

# Calculate and print averages
avg_temp = average_temperature(temperatures)
avg_humidity = average_humidity(humidity_levels)

print("\nAverage Temperature:", avg_temp, "°C")
print("Average Humidity:", avg_humidity, "%")
```
