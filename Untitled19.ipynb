```python
from abc import ABC

# Base Class
class Employee(ABC):
    def __init__(self, name, emp_id, base_salary):
        self.name = name
        self.emp_id = emp_id
        self.__base_salary = base_salary   # encapsulated (private)

    # Getter for base salary (encapsulation)
    def get_base_salary(self):
        return self.__base_salary

    @abstractmethod
    def calculate_pay(self):
        pass

    def __repr__(self):
        return f"Employee(Name: {self.name}, ID: {self.emp_id})"


# Full-Time Employee Class
class FullTimeEmployee(Employee):
    def __init__(self, name, emp_id, base_salary, bonus):
        super().__init__(name, emp_id, base_salary)
        self.__bonus = bonus   # private bonus

    # Getter for bonus
    def get_bonus(self):
        return self.__bonus

    # Polymorphic method
    def calculate_pay(self):
        return self.get_base_salary() + self.__bonus

    def __repr__(self):
        return (f"FullTimeEmployee(Name: {self.name}, ID: {self.emp_id}, "
                f"Base Salary: {self.get_base_salary()}, Bonus: {self.__bonus})")


# Part-Time Employee Class
class PartTimeEmployee(Employee):
    def __init__(self, name, emp_id, hourly_rate, hours_worked):
        super().__init__(name, emp_id, hourly_rate)   # store hourly_rate as base
        self.__hours_worked = hours_worked  # private hours

    # Getter for hours
    def get_hours_worked(self):
        return self.__hours_worked

    # Polymorphic method
    def calculate_pay(self):
        return self.get_base_salary() * self.__hours_worked

    def __repr__(self):
        return (f"PartTimeEmployee(Name: {self.name}, ID: {self.emp_id}, "
                f"Hourly Rate: {self.get_base_salary()}, Hours Worked: {self.__hours_worked})")
```

```python
emp1 = FullTimeEmployee("Alice", 101, 40000, 5000)
emp2 = PartTimeEmployee("Bob", 102, 500, 40)

print(emp1)
print("Full-Time Pay:", emp1.calculate_pay())

print()

print(emp2)
print("Part-Time Pay:", emp2.calculate_pay())
```
