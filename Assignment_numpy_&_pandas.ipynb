NUMPY PROBLEMS

```python
import numpy as np
import pandas as pd

# Load dataset
df = pd.read_csv("auto-mpg (1).csv")

# Display basic info
df.head()
```

```python
mpg = df['mpg'].to_numpy()

mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_mpg = np.std(mpg)

cars_above_25 = np.sum(mpg > 25)

mean_mpg, median_mpg, std_mpg, cars_above_25
```

```python
cylinders = df['cylinders'].to_numpy()
car_names = df['car name'].to_numpy()

filtered_cars = car_names[cylinders > 6].tolist()
filtered_cars
```

```python
weight = df['weight'].to_numpy()

np.percentile(weight, [25, 50, 75])
```

```python
acc = df['acceleration'].to_numpy()

acc_normalized = (acc - acc.min()) / (acc.max() - acc.min())
acc_normalized
```

```python
hp = pd.to_numeric(df['horsepower'], errors='coerce').to_numpy()

hp_mean = np.nanmean(hp)
hp_filled = np.where(np.isnan(hp), hp_mean, hp)

hp_increased = hp_filled * 1.10
hp_increased
```

```python
displacement = df['displacement'].to_numpy()
origin = df['origin'].to_numpy()

avg_displacement_europe = displacement[origin == 2].mean()
avg_displacement_europe
```

```python
matrix = df[['mpg', 'horsepower', 'weight']].apply(
    pd.to_numeric, errors='coerce'
).fillna(0).to_numpy()

vector = np.array([1, 0.5, -0.2])

dot_product = matrix.dot(vector)
dot_product
```

PANDAS PROBLEMS

```python
df.head(10)
df.shape
df.describe()
```

```python
df_1975 = df[(df['model_year'] == 75) & (df['weight'] < 3000)]
df_1975[['car_name', 'weight', 'mpg']]
```

```python
df.isnull().sum()

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)
```

```python
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
```

```python
df.groupby('origin')['mpg'].mean()
```

```python
df.sort_values(by='mpg', ascending=False).head(10)
```

```python
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)
```

```python
summary_df = df.groupby('model_year')[['mpg', 'weight', 'horsepower']].mean()
summary_df
```

```python
high_mpg = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg.to_csv("high_mpg_cars.csv", index=False)
```

```python
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[
    (df['mpg'] < Q1 - 1.5 * IQR) |
    (df['mpg'] > Q3 + 1.5 * IQR)
]

outliers[['car_name', 'mpg', 'model_year']]
```
